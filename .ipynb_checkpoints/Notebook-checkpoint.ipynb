## Functions and imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os, sys
import copy
import pickle

plt.rcParams['font.size'] = 28
plt.style.use('ggplot')
plt.rcParams["axes.grid"] = False
c = plt.rcParams['axes.prop_cycle'].by_key()['color']

import keras
from keras.models import Model, load_model
from keras.layers import Activation, Input, Flatten, Masking, RepeatVector, \
                         TimeDistributed, Dense, Concatenate, Dropout, LSTM, \
                         GRU, SimpleRNN, Bidirectional, Embedding, BatchNormalization
from keras.optimizers import Adam
from keras.utils import np_utils, to_categorical
from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import ModelCheckpoint, EarlyStopping


def getRawDataFromFile(datapath="data/personachat/", file="train_both_revised.txt"):
    
    f = open(datapath+file)

    conversations = []
    current_conversation = []
    
    for l, line in enumerate(f):
        #print(l, line)
        if "persona:" in line:
            if len(current_conversation) > 1:
                conversations.append(current_conversation)
            current_conversation = [] 
            continue

        #remove numberings
        processed_line = line.split(' ')
        processed_line = " ".join(processed_line[1:])
        line = processed_line
        #print(line)

        conv = line.split('\t')    
        q = conv[0]
        a = conv[1]
        current_conversation.append(q)
        current_conversation.append(a)
    
    return conversations 
```

### Extract word tokens

Compute the set of unique words (dictionary) in all sentences along with the number of occurences of each of these words.

```python
def dictionary_creator(conversations):
    dictionary = {}
    for con in conversations:
        for sent in con:
            words = sent.split(' ')
            for w in words:
                if w not in dictionary.keys():
                    dictionary[w] = 1
                else:
                    dictionary[w] += 1
    return dictionary
```

## Filtering

Filtering our data by removing sentences containing rare words. 

```python
N = 80
rm_words = set()
filtered_word_count = {}
for k,v in word_count.items():
    if v >= N:
        filtered_word_count[k] = v
    else:
        rm_words.add(k)
```

```python
new_convs = []
for conv in conversations:
    tmp = []
    for sent in conv:
        rm = 0
        for word in sent.split(' '):
            if word in rm_words:
                rm = 1
        if rm == 0:
            tmp.append(sent)
    new_convs.append(tmp)
```

**Comment**: Extremely rare words are kind of like outliers and do not really affect text representations. Therefore removing them can help learn a more generic context among our remaining sentences and improve our model's ability to generalize to other conversations. 

### Tokenization and padding

```python
word_int = {}
word_int['PADD'] = 0
i = 1
for w in filtered_word_count.keys():
    word_int[w] = i
    i += 1
word_int['START'] = i
word_int['END'] = i + 1

new_convs_int = []
new_convs_int_lengths = []
for conv in new_convs:
    for sent in conv:
        sent = 'START ' + sent + ' END';
        sent_int = [word_int[w] for w in sent.split(' ')]
        new_convs_int.append(sent_int)
        new_convs_int_lengths.append(len(sent_int))
```

## Building and training generative models of language

### RNN vs LSTM vs GRU 

Building, training and comparing generative models of language based on RNNs with different recurrent units (SimpleRNN, GRU and LSTM). 

The target of the network will be to approximate the word transition probabilities Pr(word[n+1]|H[n]) with H[n]=f(word[:n]) being the hidden state of the network.  

```python
#Meta-parameters
embedding_size = 128
hidden_size = 64
dropout = 0.
recurrent_dropout = 0.

batch_size = 64
epochs = 32
validation_split = 0.2

dataset_cut = -1
```

```python
I = {}
E = {}
H = {}
H2 = {}
B = {}
B2 = {}
A = {}
A2 = {}
R = {}
Y = {}
models = {}
logs = {}
```

```python
def create_model(unit_type):
    
    I[unit_type] = Input(shape=(maxlen-1,), name="input")
    E[unit_type] = Embedding(num_words, embedding_size, mask_zero=True, name="embedding")(I[unit_type])
    
    # Recurrent layer
    if unit_type == 'RNN':
        H[unit_type] = SimpleRNN(hidden_size, dropout=recurrent_dropout,
                                 return_sequences=True, name="rnn")(E[unit_type])
    elif unit_type == 'GRU':
        H[unit_type] = GRU(hidden_size, dropout=recurrent_dropout, 
                           return_sequences=True, name="gru")(E[unit_type])
    elif unit_type == 'LSTM':
        H[unit_type] = LSTM(hidden_size, dropout=recurrent_dropout, 
                            return_sequences=True, name="lstm")(E[unit_type])
    else:
        raise Exception("invalid recurrent unit!")
    
    B[unit_type] = BatchNormalization(name="batchnorm")(H[unit_type])
    A[unit_type] = Activation('sigmoid', name="activation")(B[unit_type])
        
    # Readout layer
    R[unit_type] = Dense(128, name="readout")(A[unit_type])
    B2[unit_type] = BatchNormalization(name="batchnorm2")(R[unit_type])
    A2[unit_type] = Activation('sigmoid', name="activation2")(B2[unit_type])
    
    # Output layer
    Y[unit_type] = Dense(num_words, activation='softmax', name="output")(A2[unit_type])

    models[unit_type] = Model(I[unit_type], Y[unit_type])
    models[unit_type].compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['acc'])
    models[unit_type].summary()

def train_model(unit_type):
    file_name = "saved_data/model_checkpoint" + unit_type + ".hdf5"
    checkpoint = ModelCheckpoint(file_name, monitor='val_loss', verbose=1, mode='min', save_best_only=True)
    
    logs[unit_type] = models[unit_type].fit({'input': X}, {'output': T}, 
                                epochs=epochs, validation_split=validation_split, 
                                batch_size=batch_size, callbacks=[checkpoint]).history
```

**Comment**: The word 'PADD' has value 0 in our input sequences. With 'mask_zero' set to 'True' in our network, it learns to treat all 0 values as as special 'padding' values which should be masked out. Consequently, the behavior of the embedding layer does not change when it encounters zero after the first time and simply copies the previous hidden state of the cell. This also allows us to treat input sequences as sentences with variable lengths although they are padded to have the same dimensions.

```python
# Define input and target matrices
# Target matrix T is 3D: num_samples x (maxlen - 1) x vocab_size

X = new_convs_int_padded[:,:-1]
tmp = new_convs_int_padded[:,1:]
T = to_categorical(tmp)
```

**Comment**: The inputs with batch size 64 (64 sentences) will have dimensions (64, maxlen-1), where maxlen is the common number of words in any sentence. By forcing the input to contain only the first word to the second last word in each sentence, we achieve maxlen-1 columns. The target of the network is three-dimensional with size (64, maxlen-1, vocabulary size) where vocabulary size is the total number of unique words in our (filtered) dataset. The target contains the second word to the last word of each sentence (i.e. input matrix "shifted" one position to the right), which are one-hot coded. This vector representation for each word allows the network to compute a probability distribution of the next word over all possible words in our vocabulary via the softmax function.

```python
# History plotting functions from miniproject 1
def prepare_standardplot(title, xlabel):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle(title)
    ax1.set_ylabel('categorical cross entropy')
    ax1.set_xlabel(xlabel)
    ax1.set_yscale('log')
    ax2.set_ylabel('accuracy [% correct]')
    ax2.set_xlabel(xlabel)
    return fig, ax1, ax2

def finalize_standardplot(fig, ax1, ax2):
    ax1handles, ax1labels = ax1.get_legend_handles_labels()
    if len(ax1labels) > 0:
        ax1.legend(ax1handles, ax1labels)
    ax2handles, ax2labels = ax2.get_legend_handles_labels()
    if len(ax2labels) > 0:
        ax2.legend(ax2handles, ax2labels)
    fig.tight_layout()
    plt.subplots_adjust(top=0.9)

def plot_history(history, title):
    fig, ax1, ax2 = prepare_standardplot(title, 'epoch')
    fig.set_figwidth(13)
    ax1.plot(history['loss'], label = "training")
    ax1.plot(history['val_loss'], label = "validation")
    ax2.plot(history['acc'], label = "training")
    ax2.plot(history['val_acc'], label = "validation")
    finalize_standardplot(fig, ax1, ax2)
    return fig

def compare_history(history, title1, title2):
    fig, ax = plt.subplots(1,2)
    fig.set_figwidth(15)
    fig.set_figheight(7)
    ax[0].plot(history[0]['val_loss'])
    ax[0].plot(history[2]['val_loss'])
    ax[0].plot(history[1]['val_loss'])
    ax[0].legend(["SimpleRNN", "LSTM", "GRU"])
    ax[0].set_title(title1)
    ax[0].set_ylabel("categorical cross entropy")
    ax[0].set_xlabel("epoch")
    
    ax[1].plot(history[0]['val_acc'])
    ax[1].plot(history[2]['val_acc'])
    ax[1].plot(history[1]['val_acc'])
    ax[1].legend(["SimpleRNN", "LSTM", "GRU"])
    ax[1].set_title(title2)
    ax[1].set_ylabel("accuracy")
    ax[1].set_xlabel("epoch")
    plt.show()
```

**Comments**:

* We can observe overfitting in the validation loss curves of all three recurrent units. For example, the validation loss reaches a minimum and begins to increase gradually after about epoch 14 for GRU and LSTM (maybe a few epochs later for simpleRNN). This could be due to the fact that we filtered our dataset to a great extent, reducing the total size of vocabulary from roughly 19K to 1.2K (and hence reducing the number of sentences). Since having more data can in general help counter overfitting, our relatively smaller dataset could explain the overfitting effect.

* We should find a balance in the degree to which the dataset is filtered. In our case, we could potentially benefit from filtering less to have a larger dataset, which may decrease overfitting.

* We can implement dropouts in the recurrent or dense layers. Together or alternatively, we can also use regularizers on the weights of the recurrent or dense layers to avoid overfitting.

* Smaller batches can be used to offer a regularizing effect (Wilson and Martinez, 2003) and therefore possibly counter overfitting and achieve better validation loss. Otherwise early stopping could be used to stop the training when the validation loss has not improved for a while so that we don't train too many epochs after the network begins to overfit.

### Optimal network size

Comparing the learning curves for three networks with 64 (previous exercise), 128 and 256 GRUs (single layer) and one with two hidden layers of 64 GRUs. 

```python
# Meta-parameters
embedding_size = 128
dropout = 0.
recurrent_dropout = 0.

batch_size = 64
epochs = 200
validation_split = 0.2

dataset_cut = -1

I = {}
E = {}
H = {}
H1 = {}
H2 = {}
B = {}
B2 = {}
B3 = {}
A = {}
A2 = {}
A3 = {}
R = {}
Y = {}
models = {}
logs = {}
num_words = len(word_int)
```

```python
def create_model_gru(unit_type, hidden_size, n_layers):
    
    I[unit_type] = Input(shape=(maxlen-1,), name="input")
    # Transfer learning
    GRUmodel = load_model("saved_data/GRUmodel_128_64")
    E[unit_type] = Embedding(num_words, embedding_size, 
                             weights = GRUmodel.layers[1].get_weights(),
                             mask_zero=True, name="embedding")(I[unit_type])
    
    # Recurrent layer
    if 'GRU_optnetsize' in unit_type:
        if n_layers == 1:
            H[unit_type] = GRU(hidden_size, dropout=recurrent_dropout, 
                           return_sequences=True, name="gru_optnetsize")(E[unit_type])
            B[unit_type] = BatchNormalization(name="batchnorm")(H[unit_type])
            A[unit_type] = Activation('sigmoid',name="activation")(B[unit_type])
        elif n_layers == 2:
            H1[unit_type] = GRU(hidden_size, dropout=recurrent_dropout, 
                           return_sequences=True, name="gru_optnetsize_h1")(E[unit_type])
            B3[unit_type] = BatchNormalization(name='batchnorm')(H1[unit_type])
            A3[unit_type] = Activation('sigmoid',name="activation")(B3[unit_type])
            H2[unit_type] = GRU(hidden_size, dropout=recurrent_dropout, 
                           return_sequences=True, name="gru_optnetsize_h2")(A3[unit_type])
            B[unit_type] = BatchNormalization(name="batchnorm2")(H2[unit_type])
            A[unit_type] = Activation('sigmoid',name="activation2")(B[unit_type])
        else:
            raise Exception("invalid number of hidden layers")
    else:
        raise Exception("invalid recurrent unit!")
    
    # Readout layer
    R[unit_type] = Dense(128, activation='sigmoid', name="readout")(A[unit_type])
    B2[unit_type] = BatchNormalization(name="batchnorm_readout")(R[unit_type])
    A2[unit_type] = Activation('sigmoid',name="activation_readout")(B2[unit_type])
        
    # Output layer
    Y[unit_type] = Dense(num_words, activation='softmax', name="output")(A2[unit_type])

    models[unit_type] = Model(I[unit_type], Y[unit_type])
    models[unit_type].compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['acc'])
    models[unit_type].summary()

def train_model_gru(unit_type):
    file_name = "saved_data/model_optnetsize_checkpoint_" + unit_type + "_" + str(hidden_size) + "_" + str(n_layers) + ".hdf5"    
    checkpoint = ModelCheckpoint(file_name, monitor='val_loss', verbose=1, mode='min', save_best_only=True)
    earlystopping = EarlyStopping(monitor='val_acc', min_delta=0.001, patience=2)
    
    logs[unit_type] = models[unit_type].fit({'input': X}, {'output': T}, 
                                epochs=epochs, validation_split=validation_split, 
                                batch_size=batch_size, callbacks=[checkpoint, earlystopping]).history
```

**Comments** :
* From smallest to largest model in terms of the number of parameters, we have GRU_64_64, GRU_128, and GRU_256. We notice that the larger the model, the lower the validation loss is when it begins training and when it terminates. This may imply that in our problem we could potentially benefit from a more complex model which seems to learn better.
* Similarly, the larger the model, the earlier it stops i.e. terminates after a fewer number of epochs. However, the larger model still attains a lower loss. This suggests that a more complex model learns more quickly and begins to overfit faster.

* Early stopping can potentially terminate the training process before we complete all epochs. An advantage is that we can reduce training time if we know we are going to overfit after a certain number of epochs, but we are not really sure exactly when. A drawback is that in general early stopping could force the training to stop when we are only at a local optimum, or when it was only due to random fluctuations that we haven't improved for a short while. In these cases, further epochs could reveal even better solutions.

* Our best model is the GRU model with a single layer and 256 units because it has the lowest validation loss and higher validation accuracy.

### Generate sentences

```python
def sample(preds, temperature=1.):
    # helper function to sample an index from a probability array
    if temperature == 0.:
        return np.argmax(preds)
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)

def initial_sentence(size):
    base = np.zeros(size)
    base[0] = word_int['START']
    return base.reshape((1,size))

def generate_sentences(model, t):
    tmp_collection = []
    tmp_prob_collection = []
    for n in range(10):
        gen_sent = initial_sentence(maxlen-1)
        all_probs = []
        for s in range(maxlen-1):
            # Choose only the next word's trans. probabilities
            trans_prob = model.predict(gen_sent)[0,s]
            sampled_index = sample(trans_prob, t)
            gen_sent[0,s+1] = sampled_index
            all_probs.append(trans_prob)
            if s + 1 == 20 or sampled_index == word_int['END']:
                temp = gen_sent
                gen_sent = initial_sentence(maxlen)
                gen_sent[0,1:] = temp
                if sampled_index == word_int['END']:
                    del all_probs[-1]
                break
        sentence = int_to_str(gen_sent[0,1:])
        tmp_collection.append(sentence)
        if n == 9:
            tmp_prob_collection.append(all_probs)
    return tmp_collection, tmp_prob_collection
```

Seeding our model with two different beginnings of max 4 words and let it generate 10 possible continuations (use sampling temperature of 1.).

```python
def int_to_str_v2(sent_int):
    tmp = [int_word[i] for i in sent_int]
    s = ' '.join(w for w in tmp)
    return s.split('PADD')[0]

def generate_sentences_inter(model, base, t):
    """Generates a sentence given some intermediate sentence"""
    base_int = np.zeros(len(base)+(maxlen-1-len(base)))
    for i,w in enumerate(base):
        base_int[i] = word_int[w]
    base_int = base_int.reshape((1,maxlen-1))
    
    for s in range(len(base),maxlen):
            # Choose only the next word's trans. probabilities
            trans_prob = model.predict(base_int)[0,s-1]
            sampled_index = sample(trans_prob, t)
            
            if sampled_index == word_int['END']:
                return int_to_str_v2(base_int[0,:])
            else:
                if s == maxlen-1:
                    tmp = np.zeros(maxlen).reshahpe((1,maxlen))
                    tmp[0,:-1] = base_int
                    tmp[0,-1] = sampled_index
                    base_int = tmp
                else:
                    base_int[0,s] = sampled_index
                            
    return int_to_str_v2(base_int[0,:])
```

**Comments**: 

* The higher the temperature, the more likely we are to sample words that are likely to be random, hence generating an 'incorrect' or non-sensical sentence. The generated sentence may also be longer due to this. The closer the temperature is to 0, the more greedy we become by sampling the word with the highest probability more often. This results in generating the exact same sentence over and over again given the same starting word.

* Temperature 0 means that we are never going to permit sampling from unlikely candidate words, resulting in always selecting the word with the highest probability (completely greedy approach i.e. argmax(probability distribution)).

* Temperature of 1 means that we sample words exactly according to the predicted probability distribution over all words as computed by the softmax function. So while it is most likely to sample the word with the highest probability, we can always sample other words from time to time.

* A very high temperature (infinity) corresponds to sampling from an effectively uniform distribution where all words are equally likely to be sampled. This reuslts in more diversity but also more 'mistakes'.

*  When the temperature is low, a sentence is generated by choosing the next word with the highest probability. This is the reason why at low temperatures (e.g. 0 or 0.25), the plot is diagonal and the probability becomes progressively larger as illustrated by growing color intensity. At higher temperatures, this is not necessarily true and sometimes a word with near zero probability is nevertheless chosen as the next word. As a result, the plot no longer appears diagonal and becomes progressively scattered.

* Yes, especially at relatively high temperatures (t >= 1), there can be multiple words with non-zero probability. This is particularly the case when the preceeding word is a subject such as 'i'. Then the next word can be a verb with many candidates. In general, words with similar meanings or roles in terms of language can have non-zero probability in the same time step.

## Chatbot

Construct a model with which you can chat. The network will take as input a sentence and output a response.

```python
conversations = getRawDataFromFile()
word_count = dictionary_creator(conversations)

# getting the set of words to remove
N = 10
rm_words = set()
filtered_word_count = {}
for k,v in word_count.items():
    if v >= N:
        filtered_word_count[k] = v
    else:
        rm_words.add(k)

# creating word - int dictionnary
word_int = {}
word_int['PADD'] = 0
i = 1
for w in filtered_word_count.keys():
    word_int[w] = i
    i += 1
word_int['START'] = i
word_int['END'] = i + 1

# creating pairs dataset
all_pairs = []
for conv in conversations:
    for i in range(len(conv)-1):
        all_pairs.append([conv[i], conv[i+1]])

dataset = []
for pair in all_pairs:
    rm = 0
    for word in pair[0].split(' '):
        if word in rm_words:
            rm = 1
    for word in pair[1].split(' '):
        if word in rm_words:
            rm = 1
    if rm == 0:
        dataset.append(pair)
```

```python
#Meta-parameters
embedding_size = 128
hidden_size = 256
dropout = 0.
recurrent_dropout = 0.

batch_size = 64
epochs = 32
validation_split = 0.2

dataset_cut = -1
num_words = len(word_int)
```

```python
def create_chatbot():    
    # Context
    context = Input(shape=(maxlen-2,), name="input_context")
    context_embedding = Embedding(num_words, embedding_size, mask_zero=True, name="embedding_context")(context)
    encoder_y, state_h, state_c = LSTM(hidden_size, return_state=True, name="encoder")(context_embedding)

    # Answer
    answer = Input(shape=(maxlen-1,), name="input_answer")
    answer_embedding = Embedding(num_words, embedding_size, mask_zero=True, name="embedding_answer")(answer)
    decoder, _, _ = LSTM(hidden_size,return_sequences=True, return_state=True,
                         name="decoder")(answer_embedding, initial_state=[state_h, state_c])

    # Output
    #R = TimeDistributed(Dense(128, activation='relu'), name='readout')(decoder)
    Y = TimeDistributed(Dense(num_words, activation='softmax'), name='output')(decoder)

    Chatbot = Model(inputs = [context, answer], outputs = [Y])
    Chatbot.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['acc'])
    Chatbot.summary()
    return Chatbot

def train_chatbot(Chatbot):
    file_name = "models_and_logs/model_checkpoint_chatbotv2.hdf5"
    checkpoint = ModelCheckpoint(file_name, monitor='val_loss', verbose=1, mode='min', save_best_only=True)
    earlystopping = EarlyStopping(monitor='val_acc', min_delta=0.001, patience=10)
    
    log = Chatbot.fit(X, {'output': T}, 
                                epochs=epochs, validation_split=validation_split, 
                                batch_size=batch_size, callbacks=[checkpoint, earlystopping]).history
    return log
```

Adapting our sentence generation code from before so that we can generate an answer given a context sentence from our model.

```python
#load
with open("saved_data/data_chatbot.pkl", "rb") as file:
    [maxlen, word_int, dataset0_int_padded, dataset1_int_padded] = pickle.load(file)
```

```python
int_word = {v: k for k, v in word_int.items()}
def int_to_str_v3(sent_int, is_context):
    if is_context:
        tmp = [int_word[sent_int[i]] for i in range(len(sent_int))]
        s = ' '.join(w for w in tmp)
        return s.split('PADD')[0]
    else:
        tmp = [int_word[sent_int[i]] for i in range(len(sent_int))][1:-1]
        s = ' '.join(w for w in tmp)
        return s.split('END')[0]
```

```python
def generate_sentence_chatbot(model, context_sentence, t):
    gen_sent = initial_sentence(maxlen-1)
    all_probs = []

    for s in range(maxlen-1):
        
        pred = {}
        pred['input_context'] = context_sentence.reshape((1,22))
        pred['input_answer'] = gen_sent
        trans_prob = model.predict(pred)[0,s]

        sampled_index = sample(trans_prob, t)
        gen_sent[0,s+1] = sampled_index
        all_probs.append(trans_prob)
        if s + 1 == maxlen-2 or sampled_index == word_int['END']:
            temp = gen_sent
            gen_sent = initial_sentence(maxlen)
            gen_sent[0,1:] = temp
            if sampled_index == word_int['END']:
                del all_probs[-1]
            break
    sentence = int_to_str_v3(gen_sent[0,1:], False)
    return sentence
```
